```python
import pandas as pd
from statsmodels.stats.anova import AnovaRM
pd.options.display.max_rows = 200
```

```python
def highlight_sig(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_sig = s < 0.05
    return ['background-color: yellow' if v else '' for v in is_sig]
```

```python
df = pd.read_csv('elec81p400.txt', sep='\t')
```

```python
df
```

```python
df["ERPset"].is_unique
```

```python
melt_df = df.melt(
    id_vars=['ERPset'], 
    value_vars = ['bin1_HowFace_E81','bin2_HowHand_E81', 'bin3_WhyFace_E81', 'bin4_WhyHand_E81'],
    var_name='condition',
    value_name = 'amplitude',
)
```

```python
melt_df.sort_values(by=['ERPset','condition'], ascending=True)
```

```python
aovrm = AnovaRM(melt_df, 'amplitude', 'ERPset', within=['condition'])
res = aovrm.fit()
```

```python
from scipy.stats import ttest_ind
```

```python
conditions = ['bin1_HowFace_E81', 'bin2_HowHand_E81', 'bin3_WhyFace_E81', 'bin4_WhyHand_E81']
ttest_df = pd.DataFrame()
 
for condition in conditions:
    col = []
    for sub in conditions:
        col.append(ttest_ind(df[condition], df[sub]).pvalue)
    ttest_df[condition] = col

ttest_df = ttest_df.rename(index=dict(zip([0,1,2,3,4], conditions)))
ttest_df = ttest_df.style.apply(highlight_sig)
```

```python
ttest_df
```

```python

def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
ttest_df.style.apply(highlight_max)
```
